Visualize the probability $P(\text{Infected} | \text{Pos})$ that a person is really infected, given that they had a positive test, as a function of specificity of the test and the infection prevalence.

Infection prevalence: takes any real-number value between 0.001% to 50%.

Specificity: takes values 99%, 99.9%,99.99%, and 99.999%

Sensitivity: 99%

We can derive the following formula for $P(\text{Infected} | \text{Pos})$:

$$ 
P(\text{Infected} | \text{Pos}) = \frac{\text{sensitivity} \cdot \text{prevalence}}{\text{sensitivity} \cdot \text{prevalence} + (100 - \text{specificity}) \cdot (100 - \text{prevalence})}
$$

```python
import numpy as np
import matplotlib.pyplot as plt


# for a given specificity, sensitivity, and prevalence, calculate the probability that a person is infected, given that they had a positive test
# using the formula we derived above
def prob_infected_given_pos(spec, prev, sens):
    prob = sens * prev / (sens * prev + (100 - spec) * (100 - prev))
    return prob
```

```python
specificity = [99, 99.9, 99.99, 99.999]

sensitivity = 99

prevalence = np.logspace(-3, np.log10(50), 100)

# Plot
plt.figure(figsize=(8, 5))


for spec in specificity:
    P_infected_given_pos = prob_infected_given_pos(spec, prevalence, sensitivity)
    plt.plot(prevalence, P_infected_given_pos, label=f"Specificity = {spec}%")

# use log-scale for prevalence to better distinguish what happens for small values
plt.xscale("log")

# Plot labels and title
plt.xlabel("Prevalence (%)")
plt.ylabel("P(Infected | Positive)")
plt.title(f"P(Infected | Positive) vs. Prevalence (Specificity = {spec}%)")
plt.legend() 
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```
